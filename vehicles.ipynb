```python
#defining some required dataframes, series, arrays and dictionary 
state=df["State/ Union Territory"]
cat=df["Category"]
month=[df["January"],df["February"],df["March"],df["April"],df["May"],df["June"],df["July"],df["August"],df["September"],
       df["October"],df["November"],df["December"]]
name=["January","February","March","April","May","June","July","August","September","October","November","December"]
x=np.array(df["Total"])
y=np.array(df["State/ Union Territory"])
all=df["Total"]
mean=df["Mean"]
co=["aqua","#66CDAA","#E3CF57","blue","red","#FFD39B","cyan","#FF6103","#8B4513","#BF3EFF","#C1FFC1","#EE1289","#EE6AA7","#7CFC00"]
dict={}
for i in range(0,20):
    dict[x[i]]=y[i]
```

```python
#for increasing the size of graph
plt.rcParams["figure.figsize"]=(18,10)
```

# Defining the functions for Analysis 

```python
def ut_st(): 
    plt.hist(cat,width=0.2,color="purple")
    plt.title("UNION TERRITORIES and STATES ",fontsize=20)
    plt.xlabel("Category",fontsize=20)
    plt.ylabel("Number",fontsize=20)
```

```python
def mon_line(n):
    for i in range(1,12+1):
        if(n==i):
            mon=month[i-1]
            fig=plt.figure(1)
            ax=fig.add_subplot(111)
            ax.set_xticklabels(state,rotation=45)
            plt.plot(state,mon,lw=5,color=co[i+1])
            plt.title(name[i-1],fontsize=20)
            plt.xlabel("No. of Accidents",fontsize=20)
            plt.ylabel("States/Union Territories",fontsize=20)
            for i, j in zip(state,mon):
                plt.text(i,j,'({}, {})'.format(i,j),fontsize=12)
            plt.show()
            break
```

```python
def tot():
    fig=plt.figure(1)
    ax=fig.add_subplot(111)
    ax.set_xticklabels(state,rotation=45)
    plt.bar(state,df['Total'],color="#E3CF57")
    plt.plot(state,df['Total'],color="black")
    plt.title("TOTAL ACCIDENTS v/s STATES",fontsize=20)
    plt.xlabel("States/Union Territories",fontsize=20)
    plt.ylabel("No. of Accidents",fontsize=20)
    for i, j in zip(state,df['Total']):
                plt.text(i,j,'({}, {})'.format(i,j),fontsize=12)
```

```python
def rank():
    from collections import OrderedDict
    
    dict_i=OrderedDict(sorted(dict.items()))
    ia=1
    for key, value in dict_i.items():
        print(ia, ' : ', value)
        ia=ia+1
```

```python
def pie():
    plt.pie(df['Total'],labels=state)
    plt.title("ACCIDENT in Different STATES",fontsize=20)
    plt.show()
```

```python
def predict():
    fig=plt.figure(1)
    ax=fig.add_subplot(111)
    ax.set_xticklabels(state,rotation=45)
    plt.bar(state,mean,color="cyan")
    plt.plot(state,mean,color="black")
    plt.title("EXPECTED ACCIDENTS",fontsize=20)
    plt.xlabel("States/Union Territories",fontsize=20)
    plt.ylabel("No. of Accidents",fontsize=20)
    for i, j in zip(state,mean):
                plt.text(i,j,'({}, {})'.format(i,j),fontsize=12)
```

```python
print("\t'Month-vise Accidents' in each 'States'\n\n")
print("\n1-January\n2-February\n3-March\n4-April\n5-May\n6-June\n7-July\n8-August\n9-September\10-October\n11-November\n12-December\n0-EXIT\n")

for i in range(0,12+1):
    n=int(input("Enter your month number(eg. 1 (for jan), 0 (for exit)):"))
    if(n>12):
        print("\n\tWrong input, try again!")
    if(n==0):
        print("\n\n\t\tExit SuccesfuSuccessful!")
        break
    mon_line(n)
```

```python
print("\n\n\t 'TOTAL ACCIDENTS' in each 'STATES' ")
tot()
```

```python
print("\t'RANKS' in order of 'Less Accidents'\n\n ")
print("Rank v/s Union Territories/States\n")
rank()
pie()
```

```python
print("Predicting the Accidents for next month")
predict()
```
